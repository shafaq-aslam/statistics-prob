# Studying Central Tendency

```python
plt.figure(figsize = (20, 5))

x = range(df['RIDAGEYR'].size)
y = df['RIDAGEYR']

plt.scatter(x, y, marker='x', s=0.5)

plt.title("Data of Individual")
plt.xlabel("ID of Individual")
plt.ylabel("Data")

# plot the estimate line 
p1 = (0, 10)
p2 = (10200, 10)
x_c, y_c = zip(p1, p2)
plt.plot(x_c, y_c, color='red', linewidth=0.75)
plt.show()
```

```python
dt = df['RIDAGEYR']
dt.mean()
```

```python
plt.figure(figsize = (20, 5))

x = range(df['RIDAGEYR'].size)
y = df['RIDAGEYR']

plt.scatter(x, y, marker='x', s=0.5)

plt.title("Data of Individual")
plt.xlabel("ID of Individual")
plt.ylabel("Data")

# plot the estimate line 
x_c, y_c = ([0, 10110], [10, 10])
plt.plot(x_c, y_c, color='red', linewidth=0.75)

x_c, y_c = ([0, 10110], [dt.mean(), dt.mean()])
plt.plot(x_c, y_c, color = 'green', linewidth=2)
plt.show()
```

## Problem with Arithmetic Mean

```python
nums = np.array([3, 4, 9, 17, 18, 20, 22, 23, 29, 32, 39, 42, 42, 42, 51, 97])
```

```python
nums.mean()
```

```python
# just did sligthly change in data
nums = np.array([3, 4, 9, 17, 18, 20, 22, 23, 29, 32, 39, 42, 42, 42, 51, 200])
```

```python
# Just because of 1 number that I changed from 97 to 200, means jumped.
np.mean(nums)
```

```python
# let's try to do a different kind of average: the median! 
np.median(nums)
```
