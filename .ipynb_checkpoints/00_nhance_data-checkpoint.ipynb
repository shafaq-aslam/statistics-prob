# Data Visualization

```python
# Let's first visualize dummy small data
data = [6, 5, 8, 2, 3, 4, 9]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(data)
plt.show()
```

```python
#create a figure
plt.figure(figsize=(10, 5))

#add a line renderer with legend and line thickness
plt.plot(data)

#also need to make the figure readable--- very important for stats perspective
plt.title("First Example Plot")
plt.xlabel("x values (s)")
plt.ylabel("y values (m)")
plt.grid()

plt.show()
```

```python
# plotting ages
#create a figure
plt.figure(figsize=(20, 5))

# to add index/number
x = range(df['RIDAGEYR'].size)
# values/ages
y = df['RIDAGEYR']

plt.scatter(x, y, marker='x', s=1.5)

# to make a figure readable
plt.title("Ages of individual")
plt.xlabel("ID of individual")
plt.ylabel("Age")

plt.show()
```

```python
# to compute actual average
# sum(df['RIDAGEYR'])

# if you get nan, because of null values in data so can do this
y = df['RIDAGEYR'].dropna()
```
