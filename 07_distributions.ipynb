# Distributions

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(color_codes = True)

import datetime
```

## Bernoulli Discrete Distribution

```python
from scipy.stats import bernoulli
```

```python
data_bern = bernoulli.rvs(size = 10000, p = 0.6)
ax = sns.displot(data_bern, bins=100, kde=False)
ax.set(xlabel = "Bernoulli Distribution", ylabel = "Frequency")
```

## Binomial Discrete Distribution

```python
from scipy.stats import binom
```

```python
data_binom = binom.rvs(n = 10, p = 0.6, size = 1000)
ax = sns.displot(data_binom, bins=100, kde=False)
ax.set(xlabel = "Binomial Discrete Distribution", ylabel = "Frequency")
```
