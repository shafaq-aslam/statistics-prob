# Continuous Distributions

## Uniform Distribution

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(color_codes = True)

import datetime
```

```python
from scipy.stats import uniform
```

```python
n = 10000
# n *= 100

start = 10
width = 20

data_uniform = uniform.rvs(size = n, loc = start, scale = width)

ax = sns.displot(data_uniform, bins=100, kde=False)
ax.set(xlabel = "Uniform Distribution", ylabel = "Frequency")

# this is still discrete distribution
```

## Normal Distribution / Gaussian Distribution

```python
from scipy.stats import norm
```

```python
data_norm = norm.rvs(size=10000, loc=0, scale=1)
ax = sns.displot(data_norm, bins=100, kde=True)
plt.xlim(-10, 10) # to define limits
ax.set(xlabel="Normal Distribution", ylabel="Frequency")
```

```python
data_norm = norm.rvs(size=10000, loc=2, scale=2)
ax = sns.displot(data_norm, bins=100, kde=True)
plt.xlim(-10, 10) # to define limits
ax.set(xlabel="Normal Distribution", ylabel="Frequency")
```
