## Generating a LOT of numbers

```python
num_samples = 10000
```

```python
uniform_values = np.random.uniform(low=0.0, high=1.0, size=num_samples)
```

```python
sns.displot(uniform_values, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title('Uniform Values')
sns.despine(offset=10, trim=True) #will move axes away
plt.show()
```

```python
normal_values = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
```

```python
sns.displot(normal_values, bins=20, kde=False)
plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("Normal Values")
sns.despine(offset=10, trim=True)
plt.show()
```

```python
print(f"Mean of Uniform Values: {np.mean(uniform_values)}")
print(f"Mean of Normal Values: {np.mean(normal_values)}")
```

```python
print(f"Variance of Uniform Values: {np.var(uniform_values)}")
print(f"Variance of Normal Values: {np.var(normal_values)}")
```

```python
print(f"Standard Deviation of Uniform Values: {np.std(uniform_values)}")
print(f"Standard Deviation of Normal Values: {np.std(normal_values)}")
```

```python
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)

nv_mean = np.mean(normal_vals)
nv_std = np.std(normal_vals)
```

```python
sns.displot(normal_vals, bins=20, kde=False)
plt.xlim(-60, 60)

plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("Normal Values")
sns.despine(offset=10, trim=True)

# Ploting SD line
x_c, y_c = ([nv_mean, nv_mean],[0, 1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([nv_mean + nv_std, nv_mean + nv_std], [0, 1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

x_c, y_c = ([nv_mean - nv_std, nv_mean - nv_std], [0, 1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

plt.show()
```

```python
normal_vals_2 = np.random.normal(loc=5.0, scale=15.0, size=num_samples)

nv_mean_2 = np.mean(normal_vals)
nv_std_2 = np.std(normal_vals)
```

```python
sns.displot(normal_vals_2, bins=20, kde=False)
plt.xlim(-60, 60)

plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("Normal Values 2")
sns.despine(offset=10, trim=True)

# Ploting SD line
x_c, y_c = ([nv_mean_2, nv_mean_2],[0, 1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([nv_mean_2 + nv_std_2, nv_mean_2 + nv_std_2], [0, 1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

x_c, y_c = ([nv_mean_2 - nv_std_2, nv_mean_2 - nv_std_2], [0, 1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

plt.show()
```

```python
# Let's put both together
normal_vals_1 = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
normal_vals_2 = np.random.normal(loc=5.0, scale=15.0, size=num_samples)

def plot_dist(vals, label):
    nv_mean = np.mean(vals)
    nv_std = np.std(vals)

    sns.displot(vals, bins=20, kde=False)
    plt.xlim(-60, 60)

    plt.xlabel('Frequency')
    plt.ylabel('Values')
    plt.title(label)
    sns.despine(offset=10, trim=True);

    x_c, y_c= [(nv_mean, nv_mean), (0, 1500)]
    plt.plot(x_c, y_c, color="green", linewidth=2)

    x_c, y_c = ([nv_mean + nv_std, nv_mean + nv_std], (0, 1500))
    plt.plot(x_c, y_c, color="red", linewidth=2)

    x_c, y_c = ([nv_mean - nv_std, nv_mean - nv_std], (0, 1500))
    plt.plot(x_c, y_c, color="red", linewidth=2)

    plt.show()

plot_dist(normal_vals_1, "Less Variant")
plot_dist(normal_vals_2, "More Variant")
```

```python
med_inc = [x[0] for x in houses.data]
mean = np.mean(med_inc)
std = np.std(med_inc)

sns.displot(med_inc, bins = 20, kde=False)
# plt.xlim(-60, 60)

plt.title("Histogram of Median Incomes of California Houses")
plt.xlabel("Median Incomes")
plt.ylabel("Frequency")
sns.despine(offset=10, trim=True);

x_c, y_c = ([mean, mean], [0, 4000])
plt.plot(x_c, y_c, color="green", linewidth=2)

x_c, y_c = ([mean + std, mean + std], [0, 4000])
plt.plot(x_c, y_c, color="red", linewidth=2)

x_c, y_c = ([mean - std, mean - std], [0, 4000])
plt.plot(x_c, y_c, color="red", linewidth=2)

plt.show()
```
