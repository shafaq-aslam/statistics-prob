# One Hot Encoding

```python
B = ['bird', 'cat', 'dog', 'cat', 'bird', 'bird']
d = {'Categorical': B}
df = pd.DataFrame(d)
```

```python
df
```

```python
# "dummies" is used to create columns corresponding to unique values
dfDummies = pd.get_dummies(df['Categorical'], prefix= 'category', dtype=int)
```

```python
dfDummies 
```

```python
dfDummies.values # one hot representation
```
